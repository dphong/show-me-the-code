```python
import matplotlib.pyplot as plt # plt 用于显示图片
import matplotlib.image as mpimg # mpimg 用于读取图片
import numpy as np

lena = mpimg.imread('avatar.jpg') # 读取和代码处于同一目录下的 avatar.jpg
# 此时 lena 就已经是一个 np.array 了，可以对它进行任意处理
```

```python
lena.shape #(690, 690, 3)
```

```python
def showImg(lena, s):
    plt.imshow(lena) # 显示图片
    plt.axis('off') # 不显示坐标轴
    plt.text(x=620, y=65, s=s, fontsize=16, color="red")
#     plt.Circle((620,65),25,color='r',alpha=0.3)  #创建一个椭圆，参数：中心点，半径，默认这个圆形会跟随窗口大小进行长宽压缩
    plt.show()
```

```python
showImg(lena, "1")
```

```python
showImg(lena, "15")
```
