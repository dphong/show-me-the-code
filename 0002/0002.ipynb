```python
import pymysql
import time

# Mysql数据库连接
db = pymysql.connect(host="172.17.0.1",user="jupyter",passwd="ydiiMCASzFDinztn",port=3306,db="jupyter",use_unicode=True,charset="utf8")

# 删除同名数据表
sql = "DROP TABLE IF EXISTS `coupon_code`;"
db.query(sql)

# 创建表
sql = "CREATE TABLE `coupon_code`( \
        `id` int(10) UNSIGNED NOT NULL AUTO_INCREMENT, \
        `code` varchar(32) CHARACTER SET utf8 COLLATE utf8_general_ci NOT NULL DEFAULT '' COMMENT 'coupon code', \
        `createdTime` int(10) NOT NULL COMMENT 'timestamp', \
        PRIMARY KEY (`id`) USING BTREE \
      ) ENGINE = InnoDB AUTO_INCREMENT = 1 CHARACTER SET = utf8 COLLATE = utf8_general_ci ROW_FORMAT = Dynamic;"
db.query(sql)
```

```python
# 使用 cursor() 方法创建一个游标对象 cursor
cursor = db.cursor()

for code in codes:
    sql = "INSERT INTO coupon_code (`code`, `createdTime`) values('%s', %d)" % (code , int(time.time()))
    sql.encode('utf-8')
    cursor.execute(sql)

# 提交数据到数据库
db.commit()
# 关闭数据库连接
db.close()
```
