#### 查询出MySQL中的优惠码记录

```python
import numpy as np
import pymysql
import time

# Mysql数据库连接
db = pymysql.connect(host="172.17.0.1",user="jupyter",passwd="ydiiMCASzFDinztn",port=3306,db="jupyter",use_unicode=True,charset="utf8")

# 使用 cursor() 方法创建一个游标对象 cursor
cursor = db.cursor()

sql = 'select id, code from coupon_code Limit 200'
sql.encode('utf-8')

cursor.execute(sql)
codes = cursor.fetchall()

# 关闭数据库连接`
db.close()
```

#### 方式一：将数据插入Redis中，数据库中的 id 作为 key，code 作为 value

```python
import redis

r = redis.Redis(host='172.17.0.1', port=6379,db=0)

for code in codes:
    r.set(code[0], code[1])
```

#### 取出存储的值

```python
redisCode = []
for code in codes:
    redisCode.append([code[0], r.get(code[0])])
```

#### 方式二：将数据插入Redis中，数据库中的 code 作为 key，是否使用 作为 value，0 - 未使用，1 - 已使用

```python
import redis

r = redis.Redis(host='172.17.0.1', port=6379,db=0)

for code in codes:
    r.set(code[1], 0)
```

#### 取出存储的值

```python
redisCode = []
for code in codes:
    redisCode.append([code[1], r.get(code[1])])
```
