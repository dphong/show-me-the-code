#### 将数据插入Redis中，数据库中的 id 作为 key，code 作为 value

```python
import redis

r = redis.Redis(host='172.17.0.1', port=6379,db=0)

for code in codes:
    r.set(code[0], code[1])
```

#### 取出存储的值

```python
redisCode = []
for code in codes:
    redisCode.append([code[0], r.get(code[0])])
```
